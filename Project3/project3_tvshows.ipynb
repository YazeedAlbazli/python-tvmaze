## Let's Begin!

Establishing the API Connection

```python
from tvmaze.api import Api
api = Api()
```

```python

shows=[]
shows_name=[]
shows_seasons=[]
shows_episodes=[]
shows_characters=[]
shows_cast=[]
shows_crew=[]
shows_network=[]
    
```

### **WARNING!**
**Activating the cell below might take up to 50-60 minutes! (Maybe longer)**

```python
for i in range(1,3001):
    shows.append(api.show.list(page=i))
```

```python
for page in shows:
    for show in page:
        shows_name.append(show.name)
        shows_seasons.append(api.show.seasons)
        shows_episodes.append(api.show.episodes)
        shows_characters.append(api.character.get)
        shows_cast.append(api.show.cast)
        shows_crew.append(api.show.crew)
        shows_network.append(api.network.get)
```
